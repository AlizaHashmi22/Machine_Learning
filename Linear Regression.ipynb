```python
#univariate analysis(one column)
plt.figure(figsize=(8,4))
plt.title('Salary Distribution')
plt.xlabel('Amount of Salary')
plt.ylabel('Density')
sns.histplot(df['Salary'],kde=True) #kde is kernel density estimation
plt.grid(True)
plt.savefig('Salary_distribution_histogram.png')
plt.show
```

```python
#bivariate analysis(two column)
plt.figure(figsize=(8,4))
plt.title('Salary vs Experience ')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
sns.scatterplot(data=df,x='YearsExperience',y='Salary')
plt.grid(True)
plt.savefig('Salary_distribution_histogram.png')
plt.show
```

```python
X=df.drop(columns=["Salary"])
y=df['Salary']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=45)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg
```

```python
reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)
y_pred
```

```python
pd.DataFrame({'y_test':y_test, 'y_pred':y_pred,'MAE':(y_test-y_pred)})
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
import numpy as np
print('MSE = ', mean_squared_error(y_test, y_pred ))
print('RMSE = ', np.sqrt(mean_squared_error(y_test, y_pred )))
print('R2 = ', r2_score(y_test, y_pred ))
```

```python
reg.predict([[25]])
```

```python
print ("Intercept value:",reg.intercept_)
print ("Slope value:",reg.coef_)
print("y= mX","+","C")
print("y=",reg.coef_,"*",+reg.intercept_)
```
